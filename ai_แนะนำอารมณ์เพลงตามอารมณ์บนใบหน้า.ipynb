```python
from deepface import DeepFace
import random

# ---------------------------
# à¸à¸³à¸«à¸™à¸”à¹€à¸à¸¥à¸‡à¸•à¸²à¸¡à¸­à¸²à¸£à¸¡à¸“à¹Œ
# ---------------------------
music_dict = {
    "happy": ["Happy - Pharrell Williams", "Uptown Funk - Bruno Mars"],
    "sad": ["Fix You - Coldplay", "Someone Like You - Adele"],
    "angry": ["Believer - Imagine Dragons", "Numb - Linkin Park"],
    "neutral": ["Perfect - Ed Sheeran", "Let It Be - The Beatles"],
    "surprise": ["Shake It Off - Taylor Swift"],
    "fear": ["Lovely - Billie Eilish"],
    "disgust": ["Creep - Radiohead"]
}

# ---------------------------
# à¹€à¸¥à¸·à¸­à¸à¸£à¸¹à¸›à¸—à¸µà¹ˆà¸­à¸±à¸›à¹‚à¸«à¸¥à¸”
# ---------------------------
image_path = list(uploaded.keys())[0]

# ---------------------------
# à¸§à¸´à¹€à¸„à¸£à¸²à¸°à¸«à¹Œà¸­à¸²à¸£à¸¡à¸“à¹Œà¸ˆà¸²à¸à¹ƒà¸šà¸«à¸™à¹‰à¸²
# ---------------------------
result = DeepFace.analyze(
    img_path=image_path,
    actions=['emotion']
)

emotion = result[0]['dominant_emotion']

# ---------------------------
# à¹€à¸¥à¸·à¸­à¸à¹€à¸à¸¥à¸‡à¸—à¸µà¹ˆà¹€à¸«à¸¡à¸²à¸°à¸ªà¸¡
# ---------------------------
song = random.choice(music_dict.get(emotion, ["à¹„à¸¡à¹ˆà¸à¸šà¹€à¸à¸¥à¸‡à¸—à¸µà¹ˆà¹€à¸«à¸¡à¸²à¸°à¸ªà¸¡"]))

# ---------------------------
# à¹à¸ªà¸”à¸‡à¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œ
# ---------------------------
print("ğŸ˜€ à¸­à¸²à¸£à¸¡à¸“à¹Œà¸—à¸µà¹ˆà¸•à¸£à¸§à¸ˆà¸à¸š:", emotion)
print("ğŸµ à¹€à¸à¸¥à¸‡à¸—à¸µà¹ˆà¹à¸™à¸°à¸™à¸³:", song)
```
